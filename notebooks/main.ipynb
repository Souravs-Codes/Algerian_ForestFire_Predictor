```python
## Independant and dependant features
X=df_copy.drop('FWI',axis=1)
Y=df_copy['FWI']
X,Y
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
##Feature Slection based on correaltion
x_train.corr()
```

```python
##Check for multicollinearilty
plt.figsize=(20,20)
corr=x_train.corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
def correlation(dataset,threshold):
    col_corr = set()
    corr_matrix =dataset.corr()
    for i in range (len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features=correlation(x_train,0.85)
```

```python
##drop features when correlation is more than 0.85
x_train.drop(corr_features,axis=1,inplace = True)
x_test.drop(corr_features,axis=1,inplace = True)
```

```python
##Feature Scalling or Standardization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
##BoxPlot to understand the effect of scaler
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=x_train)
plt.title('X_train Before Scaling')
plt.subplot(1,2,2)
sns.boxplot(data=x_train_scaled)
plt.title('X_Train After scaling')
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train_scaled,y_train)
```

```python
y_pred=model.predict(x_test_scaled)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```

## LASSO REGRESSION


```python
from sklearn.linear_model import Lasso
model_lasso=Lasso()
model_lasso.fit(x_train_scaled,y_train)
y_pred=model_lasso.predict(x_test_scaled)
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,y_pred)
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
mae,score
```

## RIDGE REGRESSION

```python
from sklearn.linear_model import Ridge
model_ridge=Ridge()
model_ridge.fit(x_train_scaled,y_train)
y_pred=model_ridge.predict(x_test_scaled)
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,y_pred)
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
mae,score
```

## ELASTICNET REGRESSION


```python
from sklearn.linear_model import ElasticNet
model_elastic=ElasticNet()
model_elastic.fit(x_train_scaled,y_train)
y_pred=model_elastic.predict(x_test_scaled)
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_test,y_pred)
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
mae,score
```
